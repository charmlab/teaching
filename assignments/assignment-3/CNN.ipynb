```python
def print_dataloader_info(loader):
    """
    Print information about a PyTorch DataLoader.

    Parameters:
    - loader: The DataLoader for train or validation data.
    """
    dataset = loader.dataset

    # If the dataset is a Subset, get the original ImageFolder dataset
    # and the selected image indices.
    if isinstance(dataset, Subset):
        image_dataset = dataset.dataset
        indices = dataset.indices
    else:
        image_dataset = dataset
        indices = range(len(dataset))

    num_images = len(dataset)
    batch_size = loader.batch_size
    class_indices = image_dataset.class_to_idx
    num_classes = len(class_indices)
    filenames = [image_dataset.samples[i][0] for i in indices]

    print(f"Number of images: {num_images}")
    print(f"Batch size: {batch_size}")
    print(f"Class indices: {class_indices}")
    print(f"Number of classes: {num_classes}")
    print(f"Number of filenames loaded: {len(filenames)}")


print("Training DataLoader Info:")
print_dataloader_info(train_loader)

print("Validation DataLoader Info:")
print_dataloader_info(validation_loader)
```
